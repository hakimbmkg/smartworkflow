```python
%%bash
set -e

echo "============================================"
echo " SmartWorkflow - Google Colab"
echo "============================================"

# --- Langkah 1: Update pip ---
echo "--- Upgrade pip ---"
pip install --upgrade pip

# --- Langkah 2: Instal TensorFlow GPU ---
echo "--- Instal TensorFlow (GPU enabled) ---"
pip install tensorflow==2.16.2 tf-keras
pip install ml_dtypes>=0.5.0
pip install jax jaxlib

# --- Langkah 3: Instal dependensi utama ---
echo "--- Instal dependensi utama ---"
pip install \
    numpy \
    obspy \
    pandas \
    matplotlib \
    pyproj \
    tqdm \
    scikit-learn \
    scipy \
    h5py

echo ""

# --- Langkah 4: Kloning repository eksternal ---
CORE_DIR="core"
PHASENET_DIR="$CORE_DIR/PhaseNet"
GAMMA_DIR="$CORE_DIR/GaMMA"

mkdir -p $CORE_DIR

echo "--- Kloning PhaseNet ---"
if [ ! -d "$PHASENET_DIR" ]; then
    git clone https://github.com/AI4EPS/PhaseNet.git $PHASENET_DIR
else
    echo "PhaseNet sudah ada, dilewati"
fi

echo "--- Kloning GaMMA ---"
if [ ! -d "$GAMMA_DIR" ]; then
    git clone https://github.com/AI4EPS/GaMMA.git $GAMMA_DIR
else
    echo "GaMMA sudah ada, dilewati"
fi

echo ""

# --- Langkah 5: Instal requirements.txt (jika ada) ---
if [ -f "requirements.txt" ]; then
    echo "--- Instal requirements.txt ---"
    pip install -r requirements.txt
fi

# --- Langkah 6: Instal PhaseNet & GaMMA (editable) ---
echo "--- Instal PhaseNet & GaMMA (editable, no deps) ---"
pip install --no-deps -e $PHASENET_DIR
pip install --no-deps -e $GAMMA_DIR

echo ""

# --- Langkah 7: Verifikasi ---
echo "--- Verifikasi run.py ---"
if [ -f "run.py" ]; then
    python run.py --help || echo "run.py gagal dijalankan"
else
    echo "PERINGATAN: run.py tidak ditemukan"
fi

echo ""
echo "============================================"
echo " Install SmartWorkflow selesai"
echo "============================================"
```
